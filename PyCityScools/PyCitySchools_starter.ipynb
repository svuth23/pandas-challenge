
## PyCity Schools Analysis

This data analysis project is focused on analyzing educational data from local government/independent schools. The goal is to    derive meaningful insights and metrics to understand school performance, student achievement, and the impact of various          factors on education.
Based on the data analyzed, we have observed:

1. independent schools have higher average scores and also higher percentage of students passing in both subjects.
2. Schools with a size of 2000 or less students have higher average scores and also higher percentage of students passing in        both subjects.
3. Schools with a budget of $630 or less also have higher average scores and higher percentage passin

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Local Government Area Summary
#Calculate the total number of schools

total_schools=school_data_complete["school_name"].nunique()


#Calculate the total number of students
total_students=school_data_complete["student_name"].count()

#Calculate the total budget
total_budget=sum(school_data_complete["budget"].unique())


#Calculate the average maths score
average_math_score = (school_data_complete["maths_score"]).mean()

#Calculate the average reading score
average_reading_score = (school_data_complete["reading_score"]).mean()

#Calculate the percentage of students with a passing maths score (50 or greater)
#used count()
passing_maths_percentage = ((school_data_complete[school_data_complete["maths_score"]>=50]["Student ID"].count())/total_students)*100

#calculate the percentage of students with a passing reading score (50 or greater)
#used len() function
passing_reading_percentage = (len(school_data_complete[school_data_complete["reading_score"]>=50])/total_students)*100

#Calculate the percentage of students who passed maths and reading (% Overall Passing)
overall_passing_percentage = ((school_data_complete[(school_data_complete["reading_score"]>=50) & (school_data_complete["maths_score"]>=50)]["Student ID"].count())/total_students)*100

#formating data
total_students_formated ='{:,.0f}'.format(total_students)
total_badget_formated='{:,.2f}'.format(total_budget)

#Create a dataframe to hold the above results
LGA_SUMMARY = {
          "total school": [total_schools],
          "total student": [total_students_formated],
          "total budget":[total_badget_formated],
           "average math score":[average_math_score],
           "average reading score":[average_reading_score],
          "% passing maths":[passing_maths_percentage],
          "% passing reading":[passing_reading_percentage],
           "% Overall Passing":[overall_passing_percentage]
          }


# PRINT summary 
LGA_SUMMARY_DF= pd.DataFrame(LGA_SUMMARY)
LGA_SUMMARY_DF
```

```python
#School Name
school_summary = school_data_complete.groupby(["school_name"])
#School Name
school_names= school_data_complete["school_name"].sort_values().unique()

#School Type
#first sort schools
school_data_sorted = school_data.sort_values(by="school_name")
#school types
school_types = school_data_sorted["type"]

#Total Students
school_total_students = list(school_summary["student_name"].count())
#print(len(school_total_students))

#Total School Budget
school_budget = list(school_summary["budget"].mean())

#Per Student Budget
school_student_budget = [i/j for i,j in zip(school_budget,school_total_students)]


#school average maths
school_average_math_score = list(school_summary["maths_score"].mean())
#print(school_average_math_score)
#Average Reading Score
school_average_reading_score = list(school_summary["reading_score"].mean())


#school passing math percentage
school_summary = school_data_complete[school_data_complete["maths_score"]>=50].groupby ("school_name")

school_math_pass_perc = [(i/j)*100 for i,j in zip(school_summary["maths_score"].count(),school_total_students)] 
#lm=len(school_math_pass_perc)#checking length
#print("lm",lm)               
#% Passing Reading percentage school
school_summary = school_data_complete[school_data_complete["reading_score"]>=50].groupby ("school_name")
school_reading_pass_perc = [(i/j)*100 for i,j in zip(school_summary["reading_score"].count(),school_total_students)]  
#lr=len(school_reading_pass_perc)
#print("lr",lr)
#% Overall Passing (The percentage of students that passed maths and reading.)
school_summary = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]
school_summary_grouped = school_summary.groupby("school_name")
overall_passing_perc = (school_summary_grouped["Student ID"].count() / school_data_complete.groupby("school_name")["Student ID"].count()) * 100
#lo=len(overall_passing_perc)
#print("lo", lo)
#create dataframe
#school_total_students = [int(value) for value in school_total_students]
#school_math_pass_perc = [float(i) for i in school_math_pass_perc]
#school_reading_pass_perc = [float(i) for i in school_reading_pass_perc]

overall_passing_perc = [float(i) for i in overall_passing_perc]

school_summary_data = {
    "School Name": school_names,
    "School Type":school_types,
    "Total Students": school_total_students,  
    "Total School Budget": school_budget,
    "Per Student Budget": school_student_budget,
    "Average Maths Score": school_average_math_score,
    "Average Reading Score": school_average_reading_score,
    "% Passing Maths": school_math_pass_perc,
    "% Passing Reading": school_reading_pass_perc,
    "% Overall Passing": overall_passing_perc
}
#school_summary_data["Total Students"] = pd.to_numeric(school_summary_data["Total Students"])

school_summary_data_df = pd.DataFrame(school_summary_data)


school_summary_data_df            
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
 
# Suppress FutureWarning
pd.options.mode.chained_assignment = None
#Create bins for segmenting and sorting data
bins = [0, 584, 629, 644, 680]

# Define name for each data that falls in each interval index(bin) 
spending_group_name = ["<$585", "$585-630", "$630-645", "$645-680"]
# Reset Index in group by "school name"
#school_summary_data_df = school_summary_data_df.reset_index()
# Add Spending Ranges by Bins
school_summary_data_df["Spending Ranges"] = pd.cut(school_summary_data_df["Per Student Budget"], bins, labels=spending_group_name)
#school_summary_data_df = school_summary_data_df.drop(["Spending Ranges(Per Student)"], axis=1)
#school_summary_data_df.head()
#avg math score

avg_spending_math = school_summary_data_df.groupby("Spending Ranges")["Average Maths Score"].mean()
#avg for reading
avg_spending_read = school_summary_data_df.groupby("Spending Ranges")["Average Reading Score"].mean()
#avg spending for %passing maths
avg_spending_pass_perc_maths = school_summary_data_df.groupby("Spending Ranges")["% Passing Maths"].mean()
#avg spending for %passing reading
avg_spending_pass_perc_read = school_summary_data_df.groupby("Spending Ranges")["% Passing Reading"].mean()
#avg spending for %passing over all
avg_spending_pass_perc_over_all = school_summary_data_df.groupby("Spending Ranges")["% Overall Passing"].mean()
#avg_spending_pass_perc_over_all.head()
#creating data frame

scores_by_school_spend = {
    "Average Math Score": round(avg_spending_math, 2),
    "Average Reading Score": round(avg_spending_read, 2),
    "% Passing Math": round(avg_spending_pass_perc_maths, 2),
    "% Passing Reading": round(avg_spending_pass_perc_read, 2),
    "% Overall Passing": round(avg_spending_pass_perc_over_all, 2)
}


spending_summary=pd.DataFrame(scores_by_school_spend)
spending_summary.head()     
```
